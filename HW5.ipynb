### Задача 1
Вы провели эксперимент c упрощением формы заказа в магазине Утконос и получили результаты по метрике конверсий в покупку. Выберите метод оценки и оцените есть ли стат.значимые различия между конверсиями в двух группах при alpha = 5%. Дайте краткие рекомендации команде:

1. Число юзеров в группах , которые заходили на сайт в период эксперимента: n1 = 15550 и n2 = 15550
1. Число юзеров в группах , которые совершили хотя бы одну покупку за период эксперимента: n1 = 164 и n2 = 228
1. Конверсии : conv1 = 1.05% conv2 = 1.47%


Посмотрим через на калькуляторе:

![Результаты вычислений на калькуляторе](task1.png)

Рассчитаем через доверительный интервал:

$$(p_2 - p_1) \pm 1.96 \cdot  \sqrt{\frac{{p_1(1-p_1)}}{n_1} + \frac{{p_2(1-p_2)}}{n_2}}$$

```python
p2 = 0.0147
p1 = 0.0105
n1 = n2 = 15550

interval = [round((p2-p1)-1.96*((p1*(1-p1)/n1) + (p2*(1-p2)/n2))**0.5, 4), round((p2-p1)+1.96*((p1*(1-p1)/n1) + (p2*(1-p2)/n2))**0.5, 4)]

print(interval)
```

И калькулятор, и интервал (поскольку интервал не проходит через 0) показывают, что имеются статистически значимые различия между двумя группами. Следовательно необходимо упростить форму заказа для увеличения конверсии.
